```python
import pandas as pd
import numpy as np
df = pd.read_csv("../data/netherlands.dat", sep="\t")

df.head()
```

```python
df.describe()
```

```python
df_new = pd.DataFrame(data=df[["id"]])
df_new["choice"] = np.where(df["choice"]%10==1, 1, 0)
df_new["business_trip"] = df["purpose"]
df_new["age41+"] = df["age"]
df_new["unemployed"] = df["employ_status"]
df_new["female"] = df["gender"]
df_new["npersons"] = df["npersons"]
df_new["train_cost"] = df["rail_cost"]
df_new["train_tt"] = df["rail_ivtt"] + df["rp_rail_ovt"]
df_new["train_xfers"] = df["rail_transfers"]
df_new["train_egress_walk"] = np.where(df["rail_egr_mode"]==1, 1, 0)
df_new["train_ingress_walk"] = np.where(df["rail_acc_mode"]==1, 1, 0)
df_new["train_1st_class"] = np.where(df["seat_status"]==1, 1, 0)
df_new["car_cost"] = df["car_cost"]
df_new["car_tt"] = df["car_ivtt"]
df_new["car_walk_time"] = df["car_walk_time"].clip(lower=0)
df_new["free_parking"] = np.where(df["car_parking_fee"]==1,1,0)
df_new["flex_arrival_time"] = np.where(df["arrival_time"]==0,1,0)
```

```python
df_new.describe()
```

```python
df.to_csv("../data/netherlands.txt", sep="\t")
```

# Data description

Variable name
- choice: the choice variable 0=car, 1=rail
- business_trip: the trip was a business trip 1=yes, 0=no
- age41+: the traveller is over 41 years old 1=yes, 0=no
- unemployed: the traveller is unemployed 1=yes, 0=no
- female: the traveller is female 1=yes, 0=no
- npersons: number of persons travelling together
- train_cost: alternative specific variable cost for train choice
- train_tt: alternative specific variable travel time for train choice
- train_xfers: alternative specific variable number of transfers for train choice
- train_egress_walk: alternative specific variable dummy traveller walked from the train station 1=yes, 0=no
- train_ingress_walk: alternative specific variable dummy traveller walked to the train station 1=yes, 0=no
- train_1st_class: alternative specific variable dummy traveller used 1st class
- car_cost: alternative specific variable cost for car choice
- car_tt: alternative specific variable travel time for car choice
- car_walk_time: alternative specific variable total walking time parking/destination
- free_parking: alternative specific variable dummy for free parking 1=yes, 0=no
- flex_arrival_time: dummy if arrival time is flexible for traveller 1=yes, 0=no
